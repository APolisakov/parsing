# Добавление остатков по магазинам

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup as bs
import re
from tqdm import trange, tqdm
```

```python
pd.set_option('display.max_colwidth', 1000)
pd.set_option('display.max_columns', 1000)
```

```python
df = pd.read_csv('csv\hi_prod.csv')
df
```

```python
def count_sity(x):
    reqx = requests.get(x)
    soup = bs(reqx.text, 'html.parser')    
    data = []
    data.append((soup.findAll('div', class_='ty-product-block__sku'))[0].text.split()[1])
    element = soup.findAll('div', class_='ty-warehouses-store__availability')
    for i in element:
        data.append(i.text.strip().replace('В наличии', '1').replace('Под заказ (1-3 дня)','0'))
    return data
```

```python
df['count_sity'] = df['soup_name'].apply(count_sity)
```

```python
t = pd.DataFrame({ 'id': df['count_sity'].str[0] ,
                  'Тирасполь__Юности_14А': df['count_sity'].str[1], 
                  'Центр_Тирасполь_25_Октября_69': df['count_sity'].str[2],
                  'Центр_Бендеры_Спортивная_5': df['count_sity'].str[3],
                  'Бендеры_Советская_5': df['count_sity'].str[4],
                  'Бендеры__Советская_8': df['count_sity'].str[5], 
                  'Слободзея_Ленина_97А': df['count_sity'].str[6],
                  'Днестровск_Строителей_18': df['count_sity'].str[7],
                  'Первомайск_Ленина_88': df['count_sity'].str[8],
                  'Григориополь_К._Маркса_181А': df['count_sity'].str[9], 
                  'Григориополь__К._Маркса_172': df['count_sity'].str[10],
                  'Дубоссары_Горького_29': df['count_sity'].str[11], 
                  'Рыбница_Вальченко_43/1': df['count_sity'].str[12], 
                  'Рыбница_Кирова_135': df['count_sity'].str[13],
                  'Каменка_Кирова_240': df['count_sity'].str[14]})
t
```

```python
df = df.merge(t, on='id')
```
