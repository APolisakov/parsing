```python
import pandas as pd
```

```python
df = pd.read_csv('csv\hi_tech_020524.csv')
```

```python
df
```

```python
def category(x):
    xx = x.split('/')
    xx = xx[3]
    return xx    
```

```python
def subcategory(x):
    xx = x.split('/')
    xx = xx[4]
    return xx   
```

```python
def subgroup(x):
    xx = x.split('/')
    xx = xx[-3]
    return xx   
```

```python
def group(x):
    xx = x.split('/')
    xx = xx[-4]
    return xx 
```

```python
df['action_sale'] = 0
```

```python
df.loc[df['old_price'] > 0,'action_sale'] = 1
```

```python
df['sale'] = (100 - ((df['price']*100)/df['old_price'])).round(2) 
```

```python
df['category'] = df['soup_name'].apply(category)
```

```python
df['subcategory'] = df['soup_name'].apply(subcategory)
```

```python
df['group'] = df['soup_name'].apply(group)
```

```python
df['subgroup'] = df['soup_name'].apply(subgroup)
```
